## Make frequency spectogram and save in labelled folders

```python
wav_paths = list(Path(wav_path).glob('*.wav'))
```

```python
spectogram_path = Path('spectograms/')
```

```python
db.name[0]
```

```python
for audio_file in db.filename[:10]:
    bird_name = db.loc[db.filename == audio_file].name.values[0]
    filename  = f'{spectogram_path}/' + f'{bird_name}/' + audio_file.replace('.wav','.png')

    print(filename)
```

```python
for audio_file in db.filename:
    bird_name = db.loc[db.filename == audio_file].name.values[0]
    samples, sample_rate = librosa.load(f'{wav_path}/{audio_file}')
    fig = plt.figure(figsize=[4.72,4.72])
    ax = fig.add_subplot(111)
    ax.axes.get_xaxis().set_visible(False)
    ax.axes.get_yaxis().set_visible(False)
    ax.set_frame_on(False)
    filename  = f'{spectogram_path}/' + f'{bird_name}/' + audio_file.replace('.wav','.png')

    S = librosa.feature.melspectrogram(y=samples, sr=sample_rate)
    librosa.display.specshow(librosa.power_to_db(S, ref=np.max))
    plt.savefig(filename, dpi=400, bbox_inches='tight',pad_inches=0)
    plt.close('all')
```

### Fourier transform of audio files

```python
# Number of samples in normalized_tone
n = len(samples)

yf = scipy.fft.rfft(samples)
xf = scipy.fft.rfftfreq(n, 1 / sample_rate)

plt.plot(xf, np.abs(yf))
plt.show()
len(xf)
```

## Train a neural net to recognise bird species:

```python
path = Path('spectograms')
```

```python
filenames = get_image_files(path)
```

```python
failed = verify_images(filenames)
failed
```

```python
failed.map(Path.unlink);
```

```python
birds = DataBlock(
    blocks=(ImageBlock, CategoryBlock), 
    get_items=get_image_files, 
    splitter=RandomSplitter(valid_pct=0.2, seed=42),
    get_y=parent_label,
    item_tfms=None)
```

```python
birds.blocks
```

```python
parent_label(filenames[999])
```

```python
dls = birds.dataloaders(path)
```
